**Reference count** will keep a track that how many number of references (variables or pointers) will point to a particular object in memory.

**getrefcount()**

**Reference counting** is used in memory management:

When reference count = 0 → object is no longer needed
Memory can be freed automatically.

import sys
sys.getrefcount(object)

getrefcount() returns one extra reference than expected as it temporarily creates an additional reference.

```python
import sys
a = [1, 2]
print(sys.getrefcount(a))
b = a
print(sys.getrefcount(a))
del(b)
print(sys.getrefcount(a))
```

```python
#SET
s1 = {10,20,30,40}
s2 = {30,40,50,60}
print(s1.union(s2))           #union
print(s1)
print(s2)
for x in s1:
  print(x)
print(s1.intersection(s2))   #intersection
print(s1)
print(s1.difference(s2))
print(s1)  #difference
print(s2.difference(s1))
print(s1.symmetric_difference(s2))  #symmetricdifference
print(s2.symmetric_difference(s1))
```

```python
s1 = {10,20,30,40}
s2 = {10,20,30,40,50,60}
print(s1.issubset(s2))
print(s2.issubset(s1))
```
